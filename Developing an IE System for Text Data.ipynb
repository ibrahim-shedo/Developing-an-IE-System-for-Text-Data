```python
import spacy
import pandas as pd
from IPython.display import display

# Load spaCy's English NLP model
nlp = spacy.load("en_core_web_sm")
```

```python
def extract_information(text):
    """
    Extract named entities and relationships from the text.
    """
    doc = nlp(text)
    entities = []
    relationships = []

    # Extract named entities
    for ent in doc.ents:
        entities.append({"Entity": ent.text, "Type": ent.label_})

    # Extract relationships (subject-verb-object)
    for token in doc:
        if token.dep_ in ("nsubj", "dobj"):  # Subject and object detection
            relationships.append({"Token": token.text, "Dependency": token.dep_, "Head": token.head.text})

    return entities, relationships
```
